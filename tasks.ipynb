## Task 1: Source the data set

Import the Iris data set from the sklearn.datasets module.

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris)
```

The function load_iris shows the data in the dataset iris. 

First, it shows rows of values of 4 attributes/variables. The number of rows corresponds to the number of targets. Next, it shows targets in 3 classes: 0,1 and 2, which correspond to target_names 'setosa', 'versicolor', and 'virginica'. 

Then, it shows the characteristics of the dataset:
- The number of targets: 150 (50 in each of the three classes)
- Number of attributes: 4 and the type of attributes: all are numeric
- What the 4 attributes represent: 
    - sepal length in cm    
    - sepal width in cm   
    - petal length in cm  
    - petal width in cm
- Class: 
    - Iris-Setosa
    - Iris-Versicolour
    - Iris-Virginica

Finally, it summarises the data: mean, max, mean, SD and class correlation of each attribute. There is no missing value. The targets are equally distributed in 3 classes. 


## Task 2: Explore the Data Structure
Print and explain the shape of the data set, the first and last 5 rows of the data, the feature names, and the target classes.

```python
import numpy as np
import pandas as pd

iris_array = pd.DataFrame(data = iris['data'], columns = iris['feature_names'])
iris_array['class'] = iris['target']
iris_array
```

## Task 3: Summarize the Data
For each feature in the dataset, calculate and display:
- mean
- minimum
- maximum
- standard deviation
- median

```python
# describe() returns a statistically description of the data in the DataFrame
# Only keep specific information using loc[]

iris_array[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']].describe().loc[['mean','min','max','std','50%']]
```

## Task 4: Visualize Features
Plot histograms for each feature using matplotlib.
Add appropriate titles and axis labels.

```python
from matplotlib import pyplot as plt

# Create the histogram with 10 groups using parameter bins

a=plt.hist(iris_array['sepal length (cm)'],bins = 10, alpha = 0.5, color = 'blue')
plt.xlabel('sepal length')
plt.ylabel('number of observations')
plt.title('Histogram: sepal length', fontweight='bold')
plt.show(a) 

b=plt.hist(iris_array['sepal width (cm)'],bins = 10, alpha = 0.5, color = 'red')
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.title('Histogram: sepal width', fontweight='bold')
plt.show(b) 

c=plt.hist(iris_array['petal length (cm)'],bins = 20, alpha = 0.5, color = 'green')
plt.xlabel('petal length')
plt.ylabel('number of observations')
plt.title('Histogram: petal length', fontweight='bold')
plt.show(c) 

d=plt.hist(iris_array['petal width (cm)'],bins = 10, alpha = 0.5, color = 'orange')
plt.xlabel('petal width')
plt.ylabel('number of observations')
plt.title('Histogram: petal width', fontweight='bold')
plt.show(d) 

```

## Task 5: Investigate Relationships
Choose any two features from the data set and create a scatter plot of them. 
Color-code the three different classes of the scatter plot points.


```python
import seaborn as sns

x=iris_array['sepal length (cm)'] # Sepal length
y=iris_array['petal length (cm)'] # Petal length

# Create the scatter plot
sns.scatterplot(x=x,y=y,data=iris_array,hue='class',style='class',palette='pastel')

# Add labels and title
plt.title('Iris sepal length x petal length', fontweight='bold')
plt.legend(title='Classes')
```

## Task 6: Analyze Relationship
Use numpy.polyfit to add a regression line to the scatter plot from Task 5.

```python
#x = iris.data[:,0]  # Sepal length
#y = iris.data[:,2]  # Petal length

# polifit calculates the coefficients of the regression line, 1 for linear regression
coef = np.polyfit(x, y, 1)
# poly1d represents the regression's equation
reg = np.poly1d(coef)

# Create the scatter plot
sns.scatterplot(x=x,y=y,data=iris_array,hue='class',style='class',palette='pastel')

# Plot the regression line
plt.plot(x, reg(x), color='red')

# Add labels and title
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.title('Iris sepal length x petal length with regression Line', fontweight='bold')
```
