## Task 2: Explore the Data Structure
Print and explain the shape of the data set, the first and last 5 rows of the data, the feature names, and the target classes.

The function DataFrame from pandas is used to create a table with data reprensented in rows and columns. The number of rows is the number of observations in the dataset. Columns represent different variables selected by user.

In this case, the table 'iri' has a regtangular shape 150 rows x 5 columns. 

- 4 first columns are 4 variables: 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)' and 'petal width (cm)'. As the variables names are long, they are replaced by shorter names: 'sep_l','sep_w','pet_l' and 'pet_w'. 
- The last column is class which contains 3 values: 0, 1 and 2 (which correspond to 'setosa', 'versicolor' and'virginica').


```python
import numpy as np
import pandas as pd

# Create a table using DataFrame()
# Table 'iri' is defined as 'data' of the iris dataset, the new variables names are used to name the columns.
iri = pd.DataFrame(data = iris['data'], columns = ['sep_l','sep_w','pet_l','pet_w'])

# Add a column 'class' with the values in the attribute 'target' in 'iri'.
iri['class'] = iris['target']

# Add column 'specie': Map numeric targets to species names using target_names
iri['specie'] = iri['class'].apply(lambda x: iris.target_names[x])

# Visualise the first and last 5 rows of 'iri'.
iri

# Ressources
# add a column 'specie' using target_names
# https://chatgpt.com/c/68149832-2c00-800c-bc66-01cb7b0f749a
```

## Task 4: Visualize Features
Plot histograms for each feature using matplotlib.
Add appropriate titles and axis labels.

Librairie `matplotlib` is used for this task. 

`plt.figure()` is used to display different histograms in different windows.

The `for` loop is used to avoid repeating the same task on different features.

For each feature in 'iri', I create a histogram which contains 10 equal-width bins in the range by using parameter `bins`. Parameters `alpha`, `color` and `edgecolor` are used to define the color of bins.

Ressources:
- [Separate histograms in different windows](https://stackoverflow.com/questions/66136669/plotting-separate-histograms-using-matplotlib)

```python
from matplotlib import pyplot as plt

for column in ('sep_l','sep_w','pet_l','pet_w'):
    # Create histograms
    plt.figure()
    plt.hist(iri[column],bins = 10, alpha = 0.5, color = 'green', edgecolor = 'black')
    
    # Add labels and title
    plt.xlabel(column)
    plt.ylabel('number of observations')
    plt.title(f'Histogram: {column}', fontweight='bold')

```

```python
# Create a figure and an axis.
fig, ax = plt.subplots()

x=iri['pet_l'] # Sepal length
y=iri['pet_w'] # Petal length

# Scatter plot.
ax.scatter(x, y, c=iri['class'])

# Labels.
ax.set_xlabel('petal length (cm)')
ax.set_ylabel('petal width (cm)')
ax.legend()
ax.grid(True)

# Ressources
# scatterplot
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html#sphx-glr-gallery-lines-bars-and-markers-scatter-with-legend-py
```

For this task, I chose 2 features Petal length for the x-axe and Petal width for the y-axe. 

Create a scatter plot of 2 chosen features by using `scatter()`. Parameter `hue` is used to represent 3 classes in 3 different colors. Parameter `palette` is the choice od colors for the plot.

## Task 6: Analyze Relationship
Use numpy.polyfit to add a regression line to the scatter plot from Task 5.

The equation of the regression is define by 2 parameters a and b: y = ax + b.

`polyfit()` in `numpy` is used to calculate the coefficients of the regression line, 1 for linear regression. In this function, `x` is defined as 'pet_l' and `y` is defined as 'pet_w'. The output is a = 0.41575542 and b = -0.36307552

```python
x=iri['pet_l'] # Petal length
y=iri['pet_w'] # Petal width

# calculate the coefficients of the regression line, 1 for linear regression
coef = np.polyfit(x, y, 1)
coef
```

Once a and b are calculated, `poly1d` is used to represent the regression's equation, which is y = 0.41575542 x - 0.36307552. 

```python
# represent the regression's equation
reg = np.poly1d(coef)
```

We now create scatter plot (of 'pet_l' and 'pet_w') and the regression line in one window. 

```python
# Create the scatter plot
sns.scatterplot(x=x,y=y,data=iri,hue='class',palette='pastel')

# Plot the regression line
plt.plot(x, reg(x), color='red')

# Add labels and title
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.title('Iris petal length x petal width with regression Line', fontweight='bold')
```

## Task 8: Compute Correlations
Calculate the correlation coefficients between the features.
Display the results as a heatmap using matplotlib.

Create the correlation matrix named 'cor' using the function `corr`. We calculate the correlations of 4 features 'sep_l','sep_w','pet_l' and 'pet_w'. We use `round()` to round the output to 2 decimals

```python
# In the table iri, 4 features are selected for correlation matrix
cor = round(iri[['sep_l','sep_w','pet_l','pet_w']].corr(),2)

# Display the correlation matrix
print(cor)
```

To visualise the correlation matrix 'cor', we can use `heatmap`

```python
sns.heatmap(cor, cmap="coolwarm", annot=True)
plt.title(f"Correlation Matrix")
plt.show()
```
