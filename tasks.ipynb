## Task 2: Explore the Data Structure
Print and explain the shape of the data set, the first and last 5 rows of the data, the feature names, and the target classes.

The function DataFrame from pandas creates a table with data reprensented in rows and columns. The number of rows is the number of observations in the dataset. Columns represent different variables selected by user.

In this case, the table iris_array has a regtangular shape 150 rows x 5 columns. 4 first columns are 4 features: sepal length, sepal width, petal length, petal width. The last column is class which contains 3 values: 0, 1 and 2. 


```python
import numpy as np
import pandas as pd

# Create a table using DataFrame()
# Table 'iri' is defined as 'data' of the iris dataset, the attribute 'feature_names' is used to name the columns.
iri = pd.DataFrame(data = iris['data'], columns = iris['feature_names'])

# Rename the feature_names to shorter names
iri.rename(columns={
    'sepal length (cm)': 'sep_l',
    'sepal width (cm)': 'sep_w',
    'petal length (cm)': 'pet_l',
    'petal width (cm)': 'pet_w',
}, inplace=True)

# Add a column named 'class' in the table 'iris_array' with the values in the attribute 'target'.
iri['class'] = iris['target']

# Add column 'specie': Map numeric targets to species names using target_names
iri['specie'] = iri['class'].apply(lambda x: iris.target_names[x])

# Visualise the first and last 5 rows of the table 'iris_array'.
iri

# Ressources
# add a column 'specie' using target_names
# https://chatgpt.com/c/68149832-2c00-800c-bc66-01cb7b0f749a
```

## Task 5: Investigate Relationships
Choose any two features from the data set and create a scatter plot of them.

```python
import seaborn as sns

x=iris_array['sepal length (cm)'] # Sepal length
y=iris_array['petal length (cm)'] # Petal length

# Create the scatter plot
sns.scatterplot(x=x,y=y,data=iris_array,hue='class',style='class',palette='pastel')

# Add labels and title
plt.title('Iris sepal length x petal length', fontweight='bold')
plt.legend(title='Classes')
```

## Task 6: Analyze Relationship
Use numpy.polyfit to add a regression line to the scatter plot from Task 5.

```python
#x = iris.data[:,0]  # Sepal length
#y = iris.data[:,2]  # Petal length

# polifit calculates the coefficients of the regression line, 1 for linear regression
coef = np.polyfit(x, y, 1)
# poly1d represents the regression's equation
reg = np.poly1d(coef)

# Create the scatter plot
sns.scatterplot(x=x,y=y,data=iris_array,hue='class',style='class',palette='pastel')

# Plot the regression line
plt.plot(x, reg(x), color='red')

# Add labels and title
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.title('Iris sepal length x petal length with regression Line', fontweight='bold')
```

## Task 8: Compute Correlations
Calculate the correlation coefficients between the features.
Display the results as a heatmap using matplotlib.

```python
# Create the correlation matrix named 'cor' using the function corr.
# In the table iris_array, only 4 features are selected
# We can use function round() to round the output to 2 decimals
cor = round(iris_array[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']].corr(),2)

# Display the correlation matrix
print(cor)
```

```python
# To visualise the correlation between variables, we can use heatmap
# seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', annot_kws=None, linewidths=0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)

sns.heatmap(cor, cmap="Greens", annot=True)
```
