## Task 1: Source the data set

Import the Iris data set from the sklearn.datasets module.

```python
from sklearn.datasets import load_iris
iris = load_iris()
print (iris)
```

The function load_iris shows the data in the data set. 

At first, it shows rows of values of 4 attributes. The number of rows corresponds to the number of targets. Below, it shows targets are classified in 3 classes 0,1 and 2, which correspond to target_names 'setosa', 'versicolor', 'virginica'. 

Further, it shows the characteristics of the data set:
- The numbers of targes: 150 (50 in each of three classes)
- Number of attributes: 4 and the type of attributes: all are numeric
- What the 4 attributes represent: 
    - sepal length in cm\n    
    - sepal width in cm\n    
    - petal length in cm\n    
    - petal width in cm\n

## Task 4: Visualize Features
Plot histograms for each feature using matplotlib.
Add appropriate titles and axis labels.

```python
from matplotlib import pyplot as plt
a=plt.hist(iris_array['sepal length (cm)'],bins = 10, alpha = 0.5, color = 'blue')
plt.xlabel('sepal length')
plt.ylabel('number of observations')
plt.title('Histogram: sepal length', fontweight='bold')
plt.show(a) 

b=plt.hist(iris_array['sepal width (cm)'],bins = 10, alpha = 0.5, color = 'red')
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.title('Histogram: sepal width', fontweight='bold')
plt.show(b) 

c=plt.hist(iris_array['petal length (cm)'],bins = 20, alpha = 0.5, color = 'green')
plt.xlabel('petal length')
plt.ylabel('number of observations')
plt.title('Histogram: petal length', fontweight='bold')
plt.show(c) 

d=plt.hist(iris_array['petal width (cm)'],bins = 10, alpha = 0.5, color = 'orange')
plt.xlabel('petal width')
plt.ylabel('number of observations')
plt.title('Histogram: petal width', fontweight='bold')
plt.show(d) 

```
